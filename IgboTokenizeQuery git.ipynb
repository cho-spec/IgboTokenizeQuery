```python
# Select a subset of the headlines for demonstration purposes
documents = df["headline"][0:30].tolist()

# Record the start time
start_time = time.time()

# Encode the documents
document_embeddings = model.encode(documents, convert_to_tensor=True)

# Record the end time
end_time = time.time()

# Calculate and print the elapsed time
elapsed_time = end_time - start_time
print(f"Time taken to encode documents: {elapsed_time} seconds")
```

## Search Function

```python
def choya_acho(query):
    query_embedding = model.encode(query, convert_to_tensor=True)
    # Calculate cosine similarity between the query and documents
    cosine_scores = util.pytorch_cos_sim(query_embedding, document_embeddings)[0]

    # Sort documents by similarity score
    results = list(zip(range(len(cosine_scores)), cosine_scores))
    sorted_results = sorted(results, key=lambda x: x[1], reverse=True)
    response = []
    for idx, score in sorted_results:
        response.append({
            "id": idx + 1,
            "score": score,
            "term": documents[idx]
        })
    return response
```
